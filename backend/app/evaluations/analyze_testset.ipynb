```python
class Solution:
    def checkDistances(self, s: str, distance: List[int]) -> bool:
        first_occurances = [-1 for _ in range(26)]

        for idx, char in enumerate(s):
            char_number = ord(char) - ord('a')
            current = first_occurances[char_number]

            if current == -1:
                first_occurances[char_number] = idx
            else:
                first_occurances[char_number] = idx - current - 1
                if first_occurances[char_number] != distance[char_number]:
                    return False
    
            print(char, char_number, first_occurances[char_number])
        
        print(first_occurances, distance)
        return True

            
```
